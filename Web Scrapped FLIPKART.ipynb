```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
site = requests.get("https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DAPPLE")
print(site)
```

```python
print(site.content)
```

```python
soup= BeautifulSoup(site.content,'html.parser')
print(soup.prettify())
```

```python
name = soup.findAll('div', attrs= {'class':'_4rR01T'})
namelist=[]
pricelist=[]
ratinglist=[]
for n in name:
    name = n.text
    namelist.append(name)
price = soup.findAll('div',attrs = {'class':'_30jeq3 _1_WHN1'})
for p in price:
    price =p.text
    pricelist.append(price)
rating = soup.findAll('div',attrs ={'class':'_3LWZlK'})
for r in rating:
    rating =r.text
    ratinglist.append(rating)
print(namelist)
print(pricelist)
print(ratinglist)
```

```python
#comparison between 2 brands
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df contains your data with a 'Brand' column

# Filter data for Brand A and Brand B
brand_a_data = df[df['Brand'] == 'Apple']
brand_b_data = df[df['Brand'] == 'SAMSUNG']

# Create a figure and axis
plt.figure(figsize=(8,3))

# Create histograms for Brand A and Brand B
sns.histplot(data=brand_a_data, x='Selling Price', bins=[5000, 10000, 15000, 20000, 25000, 30000], color='green', label='Apple')
sns.histplot(data=brand_b_data, x='Selling Price', bins=[5000, 10000, 15000, 20000, 25000, 30000], color='red', label='SAMSUNG')

# Add labels and legend
plt.xlabel('Count')
plt.ylabel('Rating')
plt.legend()

# Show the plot
plt.show()
```

```python
top_brands = df['Brand'].value_counts().head(5)

# Create a pie chart for the top 5 brands
plt.figure(figsize=(8, 8))
plt.pie(top_brands, labels=top_brands.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Top 5 Brands')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart looks like a circle
plt.show()
```
