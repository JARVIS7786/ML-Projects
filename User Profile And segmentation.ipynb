```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r'C:\Users\yuvra\Downloads\user_profiles_for_ads.csv')
```

```python
data.head(6)
```

```python
data.isnull().sum()
```

```python
#setting the aestheitc style of the plot
sns.set_style('whitegrid')
```

```python
#creating subplots for the demographic representations
fig, axes =plt.subplots(2,2 ,figsize=(18,12))
fig.suptitle('Distribution of demographic variables')
# age distribution
sns.countplot(ax=axes[0, 0], x='Age', data=data, palette='coolwarm')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].tick_params(axis='x', rotation=45)
# gender distribution
sns.countplot(ax=axes[0, 1], x='Gender', data=data, palette='coolwarm')
axes[0, 1].set_title('Gender Distribution')
sns.countplot(ax=axes[1,0],x='Education Level',data = data,palette='coolwarm')
axes[1,0].set_title('Education level Distribution')
sns.countplot(ax=axes[1,1],x='Income Level',data= data,palette='coolwarm')
axes[1,1].set_title('Income Level Distribution')
plt.show()
```
