```python
import pandas as pd
import yfinance as yf
from datetime import datetime

start_date = datetime.now() - pd.DateOffset(months=4)
end_date = datetime.now()

tickers = ['IOC.NS', 'ONGC.NS', 'BPCL.NS', 'HINDPETRO.NS','RELIANCE.NS']

df_list = []

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)
    df_list.append(data)

df = pd.concat(df_list, keys=tickers, names=['Ticker', 'Date'])
print(df.head())
```

```python
df = df.reset_index()
df.head()
```

```python
import plotly.express as px
fig = px.line(df ,x='Date',y='Close',color ='Ticker',title= "Stock Market Performance for last 3 MOnths"
            )
fig.show()
```

```python
fig = px.area(df, x='Date',y='Close',color ='Ticker',
             facet_col='Ticker',
             labels={'Date' :'Date', 'Close':'Closing Price','Ticker' :'Company'},
             title='Stock Prices for BPCL,IOC,Hindustan,Reliance,ONGC')
fig.show()
```

```python
df['MA10']= df.groupby('Ticker')['Close'].rolling(window=10).mean().reset_index(0,drop = True)
df['MA20']= df.groupby('Ticker')['Close'].rolling(window=20).mean().reset_index(0,drop = True)
```

```python
for ticker,group in df.groupby('Ticker'):
    fig=px.line(group,x='Date',y=['Close','MA10','MA20'],
    title=f"{ticker} Moving Averages")
    fig.show()
```

```python
df['Volatility'] = df.groupby('Ticker')['Close'].pct_change().rolling(window=10).std().reset_index(0,drop=True)
fig=px.line(df, x='Date',y='Volatility',
           color='Ticker',
           title = 'Volatility of All companies')
fig.show()
```

```python
BPCL = df.loc[df['Ticker'] == 'BPCL.NS',['Date','Close']].rename(columns={'Close':'BPCL'})
IOC = df.loc[df['Ticker'] == 'IOC.NS',['Date','Close']].rename(columns={'Close':'IOC.NS'})
df_corr = pd.merge(BPCL,IOC,on = 'Date')

fig=px.scatter(df_corr,x='BPCL',y='IOC.NS',
              trendline='ols',
              title='Correation between BPCL AND IOC')
fig.show()
```
